```python
#claude ai generated this function to spliting speeches into sentences and then into chunks of 4 sentences with an overlap of 1 sentence between chunks
import re

# Abbreviations to exclude from sentence splitting
ABBREVIATIONS = r'Mr|Dr|Ms|Mrs|Prof|e\.g'

def split_into_sentences(text):
    # Protect abbreviations by replacing their periods temporarily
    text = re.sub(r'\b(' + ABBREVIATIONS + r')\.\s', r'\1<PERIOD> ', text)
    
    # Split on ". " but keep the period by using lookahead
    sentences = re.split(r'(?<=\.)\s+', text)
    
    # Restore protected periods
    sentences = [s.replace('<PERIOD>', '.') for s in sentences]
    
    # Remove empty sentences
    sentences = [s.strip() for s in sentences if s.strip()]
    
    return sentences

def create_chunks(sentences, chunk_size=4, overlap=1):
    chunks = []
    step = chunk_size - overlap  # step = 3 (4 sentences, 1 overlap)
    
    for i in range(0, len(sentences), step):
        chunk = sentences[i:i + chunk_size]
        if chunk:
            chunks.append(' '.join(chunk))
    
    return chunks

# Process each row
all_chunks = []

for idx, row in df_1.iterrows():
    sentences = split_into_sentences(str(row['text']))
    chunks = create_chunks(sentences, chunk_size=4, overlap=1)
    
    for chunk_num, chunk_text in enumerate(chunks, 1):
        doc_id = f"{row['iso_code']}-{row['year']}.{chunk_num}"
        all_chunks.append({
            'doc_id': doc_id,
            'iso_code': row['iso_code'],
            'year': row['year'],
            'text': chunk_text  # chunk_number removed
        })

# Create new dataframe
chunks_df = pd.DataFrame(all_chunks)

# Save
chunks_df.to_csv('data/temp/data_chunks.csv', index=False)

print(f"Total chunks: {len(chunks_df)}")
chunks_df.head(10)
```
